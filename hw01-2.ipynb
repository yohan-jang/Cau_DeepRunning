임의의 점 설정


```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

plt.axis([0,10,-2,10])

x1=([3, 2, 4, 3, 2, 5, 8, 9, 6, 7])
y1=([5, 6, 0, 0, 2, 4, 1, 1, -1, 5])
plt.scatter(x1, y1)
plt.title('HW1 Deep_Running', fontsize=15)
plt.show()
```

```
sigma_2x = float(0)
sigma_2y = float(0)
sigma_xy = float(0)
sigma_x = float(0)
sigma_y = float(0)


for i in range(len(x1)):  
  sigma_2x += x1[i]**2
  sigma_2y += y1[i]**2
  sigma_xy += x1[i]*y1[i]
  sigma_x += x1[i]
  sigma_y += y1[i]

```

```
x = np.arange(float(-10), float(10), float(0.25))
y = np.arange(float(-10), float(10), float(0.25))
x, y = np.meshgrid(x,y)
z = ((x**2*sigma_2x)+(x*y*sigma_x)-(x*sigma_x*sigma_y)+(x*y*sigma_x)+(y**2)-(y*sigma_y)-(x*sigma_x*sigma_y)-(y*sigma_y)+(sigma_2y))/20 #cost function
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_wireframe(x, y, z, color='black')
ax.set_xlabel(r'$theta0$')
ax.set_ylabel(r'$theta1$')
ax.set_zlabel(r'$result$')
plt.show()
```

```
theta0 = float(1) #weight 1
theta1 = float(1) #weight 2
w = float(0.001)


for i in range(10000):
  theta0 = theta0 - w*(((theta0*sigma_x**2)+(theta1*sigma_x)-(sigma_x*sigma_y))/10)
  theta1 = theta1 - w*(((theta0*sigma_x)+theta1-sigma_y)/10)


x2 = np.linspace(0,10,20)
y2 = theta0*x2+theta1
plt.plot(x2,y2)
plt.show()

print(theta0)
print(theta1)
```
